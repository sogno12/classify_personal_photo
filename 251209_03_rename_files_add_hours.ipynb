```python
import os
from datetime import datetime, timedelta

# ================= 설정 구역 =================
# 1. 파일이 있는 폴더 경로 (현재 경로라면 '.' 입력)
target_folder = './photo/250426_@나고야_&지현/4일차(0429)_나고야성'

# 2. 더할 시간 (한국 시간 변환 시 9 입력)
hours_to_add = 9

# 3. 테스트 모드 (True: 변경 전후 이름만 출력 / False: 실제로 파일 이름 변경)
# ★ 처음엔 True로 두고 결과 확인 후 False로 바꾸세요!
TEST_MODE = False
# ===========================================
```

```python
# 변경된 파일 수를 세기 위한 변수
count = 0

print(f"--- 작업 시작 (테스트 모드: {TEST_MODE}) ---")

# 폴더 내 모든 파일 확인
for filename in os.listdir(target_folder):
    # 파일명과 확장자 분리 (예: 20230101_120000 .csv)
    name, ext = os.path.splitext(filename)

    try:
        # 1. 현재 파일명을 날짜 객체로 변환 (형식: YYYYMMDD_HHmmss)
        current_dt = datetime.strptime(name, "%Y%m%d_%H%M%S")

        # 2. 시간 더하기 (timedelta 사용)
        new_dt = current_dt + timedelta(hours=hours_to_add)

        # 3. 새로운 파일명 문자열로 변환
        new_name = new_dt.strftime("%Y%m%d_%H%M%S") + ext

        # 경로 합치기
        old_path = os.path.join(target_folder, filename)
        new_path = os.path.join(target_folder, new_name)

        # 결과 출력
        print(f"[대상] {filename}  -->  {new_name}")

        # 테스트 모드가 아닐 때만 실제로 이름 변경
        if not TEST_MODE:
            os.rename(old_path, new_path)
            count += 1

    except ValueError:
        # 파일명 형식이 날짜가 아닌 경우 스킵 (예: .DS_Store, 이미 변경된 파일 등)
        pass

print(f"--- 작업 완료 ---")
if not TEST_MODE:
    print(f"총 {count}개의 파일 이름이 변경되었습니다.")
else:
    print("현재는 테스트 모드입니다. 실제로 변경하려면 TEST_MODE = False 로 설정하세요.")
```
